# Model Building

```python
from sklearn.linear_model import LogisticRegression
regressor = LogisticRegression()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
X_train = encoder.fit_transform(X_train)
```

```python
# Using RFE for feature elimination
from sklearn.feature_selection import RFE
rfe = RFE(regressor, n_features_to_select= 15)
rfe = rfe.fit(X_train, y_train)
```

```python
feature_names = encoder.get_feature_names_out()

# Zipping the feature names with RFE support and ranking
feature_ranking = list(zip(feature_names, rfe.support_, rfe.ranking_))

# Printing the feature ranking
print("Feature Ranking:")
for i in feature_ranking:
    print(i)
```

```python
feature_names = encoder.get_feature_names_out()
selected_features = [feature_names[i] for i, selected in enumerate(rfe.support_) if selected]
```

```python
selected_feature_indices = [i for i, selected in enumerate(rfe.support_) if selected]
selected_feature_names = [feature_names[i] for i in selected_feature_indices]
X_train_rfe = X_train[:, selected_feature_indices]
```

# Model 1

```python
import statsmodels.api as sm
```

```python
X_train_rfe_dense = X_train_rfe.toarray()
X_train_rfe_with_const = sm.add_constant(X_train_rfe_dense)
logistic_model = sm.GLM(y_train, X_train_rfe_with_const, family=sm.families.Binomial())
logistic_model_results = logistic_model.fit()
logistic_model_results.summary()
```

# Making predictions on train dataset.

```python
train_predictions = logistic_model_results_filtered.predict(X_train_rfe_with_const_filtered)
train_predictions_binary = np.where(train_predictions >= 0.5, 1, 0)
train_results = pd.DataFrame({'Actual': y_train, 'Predicted': train_predictions_binary})
print(train_results.head())
```

# Calculating Metrics

```python
from sklearn.metrics import roc_curve
y_train_pred_probs = logistic_model_results_filtered.predict(X_train_rfe_with_const_filtered)
fpr, tpr, thresholds = roc_curve(y_train, y_train_pred_probs)
# Finding the optimal cutoff point based on the ROC curve
optimal_idx = np.argmax(tpr - fpr)
optimal_threshold = thresholds[optimal_idx]

print("Optimal Cutoff Point:", optimal_threshold)
```

```python
y_train_pred_binary = np.where(y_train_pred_probs >= optimal_threshold, 1, 0)

# Calculating the confusion matrix
from sklearn.metrics import confusion_matrix
conf_matrix = confusion_matrix(y_train, y_train_pred_binary)

tn, fp, fn, tp = conf_matrix.ravel()
# Calculating accuracy
accuracy = (tp + tn) / (tp + tn + fp + fn)
# Calculating sensitivity (true positive rate or recall)
sensitivity = tp / (tp + fn)
# Calculating specificity (true negative rate)
specificity = tn / (tn + fp)
# Calculating precision
precision = tp / (tp + fp)
# Calculating F1 score
f1_score = 2 * (precision * sensitivity) / (precision + sensitivity)
print("Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Sensitivity: {sensitivity:.4f}")
print(f"Specificity: {specificity:.4f}")
print(f"Precision: {precision:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```

# ROC curve with Optimal Cutoff Point

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve')
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.scatter(fpr[optimal_idx], tpr[optimal_idx], color='red', label=f'Optimal Cutoff (Threshold = {optimal_threshold:.3f})')
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.grid(True)
plt.show()
```
